```python
score_cols = ['math_score', 'reading_score', 'writing_score']
df_melted = df.melt(value_vars=score_cols, var_name='Score_Type', value_name='Score')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Score_Type', y='Score', data=df_melted, palette='viridis')
plt.title('Distribution of Math, Reading, and Writing Scores')
plt.xlabel('Score Type')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Bivariate Analysis

#### 1. Compare average scores by gender

```python
gender_avg_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
gender_avg_scores_melted = gender_avg_scores.melt(id_vars='gender', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='Average_Score', hue='Score_Type', data=gender_avg_scores_melted, palette='pastel')
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
lunch_avg_scores = df.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
lunch_avg_scores_melted = lunch_avg_scores.melt(id_vars='lunch', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='lunch', y='Average_Score', hue='Score_Type', data=lunch_avg_scores_melted, palette='bright')
plt.title('Average Scores by lunch type')
plt.xlabel('lunch Type')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 2. Compare scores based on test preparation course

```python
prep_course_avg_scores = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
prep_course_avg_scores_melted = prep_course_avg_scores.melt(id_vars='test_preparation_course', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='test_preparation_course', y='Average_Score', hue='Score_Type', data=prep_course_avg_scores_melted, palette='viridis')
plt.title('Average Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 3. Relationship between reading and writing scores

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='reading_score', y='writing_score', data=df, hue='gender', palette='coolwarm', s=50, alpha=0.7)
plt.title('Relationship between Reading and Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### 4. Relationship between parental education and average score

```python
parental_edu_avg_scores = df.groupby('parental_level_of_education')[['math_score', 'reading_score', 'writing_score', 'average_score']].mean().reset_index()
parental_edu_avg_scores_melted = parental_edu_avg_scores.melt(id_vars='parental_level_of_education', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(12, 7))
sns.barplot(x='parental_level_of_education', y='Average_Score', hue='Score_Type', data=parental_edu_avg_scores_melted, palette='muted')
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 5. Correlation Heatmap of Scores

```python
score_correlation = df[['math_score', 'reading_score', 'writing_score', 'average_score']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(score_correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Scores')
plt.show()
```

#### 6. Pairplot of Scores

```python
numerical_scores = ['math_score', 'reading_score', 'writing_score', 'average_score']
sns.pairplot(df[numerical_scores], diag_kind='kde', palette='magma')
plt.suptitle('Pairplot of Scores', y=1.02)
plt.show()
```
