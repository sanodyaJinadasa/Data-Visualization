```python
score_cols = ['math_score', 'reading_score', 'writing_score']
df_melted = df.melt(value_vars=score_cols, var_name='Score_Type', value_name='Score')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Score_Type', y='Score', data=df_melted, palette='viridis')
plt.title('Distribution of Math, Reading, and Writing Scores')
plt.xlabel('Score Type')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Bivariate Analysis

#### 1. Compare average scores by gender

```python
gender_avg_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
gender_avg_scores_melted = gender_avg_scores.melt(id_vars='gender', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='Average_Score', hue='Score_Type', data=gender_avg_scores_melted, palette='viridis')
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
gender_avg_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
gender_avg_scores_melted = gender_avg_scores.melt(id_vars='gender', var_name='Score_Type', value_name='Average_Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='Average_Score', hue='Score_Type', data=gender_avg_scores_melted, palette='viridis')
plt.title('Average Scores by lunch type')
plt.xlabel('lunch Type')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
