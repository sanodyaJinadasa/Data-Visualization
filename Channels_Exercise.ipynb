```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('diamonds')
df
```

## Task 1

```python
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
```

```python
ax=sns.barplot(
    data = df,
    x='cut',
    y='price',
    palette="Blues",
    legend=False
)

ax.set_title("Diamond Price by Cut")
ax.set_ylabel("Total Sales($)")
ax.set_xlabel("Drink Type")

plt.tight_layout()
plt.show()
```

```python
ax=sns.barplot(
    data = df,
    x='cut',
    y='price',
    palette="magma",
    legend=False
)

ax.set_title("Diamond Price by Cut")
ax.set_ylabel("Total Sales($)")
ax.set_xlabel("Drink Type")

plt.tight_layout()
plt.show()
```

```python
palette = sns.color_palette("Blues", n_colors=5)

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='cut',
    y='price',
    order=cut_order,
    palette=palette
)

plt.title('Diamond Price Distribution by Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Price')
plt.show()
```

## Task 2

```python
avg_price = (
    df.groupby('cut')['price']
    .mean()
    .reindex(cut_order)
)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(avg_price.values, labels=avg_price.index,
       startangle=90, colors=sns.color_palette('pastel'))
ax.set_title('Average Diamond Price by Cut Quality', fontsize=14)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(
    avg_price.index,
    avg_price.values,
    color='steelblue'
)

ax.set_title('Average Diamond Price by Cut')
ax.set_xlabel('Cut Quality')
ax.set_ylabel('Average Price ($)')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(avg_price.values, labels=avg_price.index, autopct='%1.1f%%',
       startangle=90, colors=sns.color_palette('pastel'))
ax.set_title('Average Diamond Price by Cut Quality', fontsize=14)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(avg_price.index, avg_price.values, color='steelblue', edgecolor='black')
ax.bar_label(bars, fmt='%.2f', padding=3)
ax.set_title('Average Diamond Price by Cut Quality', fontsize=14)
ax.set_xlabel('Cut Quality', fontsize=12)
ax.set_ylabel('Average Price (USD)', fontsize=12)
plt.show()
```

```python
clarity_order = [
    'I1', 'SI2', 'SI1',
    'VS2', 'VS1',
    'VVS2', 'VVS1',
    'IF'
]
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

sns.scatterplot(
    data=df,
    x='carat',
    y='price',
    hue='clarity',
    hue_order=clarity_order,
    palette='viridis',
    alpha=0.6,
    ax=ax
)

ax.set_title('Diamond Price vs Carat Weight by Clarity')
ax.set_xlabel('Carat Weight')
ax.set_ylabel('Price ($)')
ax.legend(title='Clarity', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
