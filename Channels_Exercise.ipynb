```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('diamonds')
df
```

## Task 1

```python
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
```

```python
ax=sns.barplot(
    data = df,
    x='cut',
    y='price',
    palette="Blues",
    legend=False
)

ax.set_title("Diamond Price by Cut")
ax.set_ylabel("Total Sales($)")
ax.set_xlabel("Drink Type")

plt.tight_layout()
plt.show()
```

```python
ax=sns.barplot(
    data = df,
    x='cut',
    y='price',
    palette="magma",
    legend=False
)

ax.set_title("Diamond Price by Cut")
ax.set_ylabel("Total Sales($)")
ax.set_xlabel("Drink Type")

plt.tight_layout()
plt.show()
```

```python
palette = sns.color_palette("Blues", n_colors=5)

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='cut',
    y='price',
    order=cut_order,
    palette=palette
)

plt.title('Diamond Price Distribution by Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Price')
plt.show()
```
