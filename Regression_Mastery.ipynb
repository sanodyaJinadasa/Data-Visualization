## Data Visualization & Regression Mastery

#### Restaurant Tipping Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

The Baseline Regression

```python
plt.figure(figsize=(6,4))
sns.regplot(x="total_bill", y="tip", data=tips, color="blue",line_kws={"color":"red"})
plt.title("Total Bill vs Tip (Baseline Regression)")
plt.show()
```

The Multi-Grid Slicer

```python
sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="smoker",
    col="time",
    height=5,
    aspect=1
)

plt.suptitle("Tipping Behavior by Time and Smoking Status", y=1.05)
plt.show()
```

The All-in-One Pairplot

```python
sns.pairplot(
    tips,
    hue="day",
    kind="reg"
)

plt.show()
```

From the pairplot, the relationship between total_bill and tip is clearly stronger than the relationship between party size (size) and tip.

The total_bill vs tip plots show a tight, upward linear trend with less scatter.
The size vs tip plots are more dispersed, with weaker alignment and more variability.

Conclusion: The number of people in the party (size) has a weaker relationship with the tip compared to the total_bill, which is the stronger predictor of tipping behavior.

The Custom Dashboard

```python
# Create subsets
lunch = tips[tips["time"] == "Lunch"]
dinner = tips[tips["time"] == "Dinner"]

# Create dashboard
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left plot (Lunch)
sns.regplot(
    x="total_bill",
    y="tip",
    data=lunch,
    ax=axes[0],
    color="blue"
)
axes[0].set_title("Lunch: Total Bill vs Tip")

# Right plot (Dinner)
sns.regplot(
    x="total_bill",
    y="tip",
    data=dinner,
    ax=axes[1],
    color="green"
)
axes[1].set_title("Dinner: Total Bill vs Tip")

plt.tight_layout()
plt.show()
```
