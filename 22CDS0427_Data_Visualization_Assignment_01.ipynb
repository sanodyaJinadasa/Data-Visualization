# Simple Linear Regression using Least Squares Method

### Assignmnet 01
### 22CDS0427

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
X = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([50, 55, 65, 70, 80, 90])

# Number of observations
n = len(X)

# Calculate required sums
sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_x2 = np.sum(X**2)
sum_xy = np.sum(X * Y)

# Calculate regression coefficients
beta1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
beta0 = (sum_y / n) - beta1 * (sum_x / n)

# Print coefficients
print("Slope (β1):", beta1)
print("Intercept (β0):", round(beta0, 2))

# Regression equation
print(f"Regression Line: Ŷ = {beta0:.2f} + {beta1:.2f}X")

# Predict score for 4.5 study hours
x_new = 4.5
y_pred = beta0 + beta1 * x_new

print(f"Predicted score for {x_new} study hours: {y_pred:.2f}")

# Predicted Y values for plotting
Y_hat = beta0 + beta1 * X

# Plot data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot regression line
plt.plot(X, Y_hat, color='red', linewidth=2, label='Best Fit Line')

# Labels and title
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

# Simple Linear Regression Results

## Estimated Parameters

Using the Least Squares Method, the regression parameters were estimated as follows:

- **Slope (β₁)** = 8.0  
- **Intercept (β₀)** = 40.33

## Interpretation of the Intercept (β₀)

The intercept value of **40.33** means that:

> If a student studies for **0 hours**, the predicted exam score would be approximately **40.33 marks**.
