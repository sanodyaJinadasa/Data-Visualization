```python
import matplotlib.pyplot as plt
x = range(1, 11)
markers = ['o', 's', '^', 'v', 'D', '*', '+', 'x']
for i, marker in enumerate(markers):
 plt.plot(x, [i * 2] * 10, marker=marker, linestyle='',
 label=marker) # linestyle='' → only markers, no line
plt.title("Different Matplotlib Markers")
plt.xlabel("X-axis"); plt.ylabel("Y-axis")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
# Function to add value labels above each bar
def add_labels(x, y):
  for i in range(len(x)):
    plt.text(i, y[i] + 8, y[i], ha='right') # center-align horizontally
x = ["FAGS", "FAPS", "FOC", "FGS", "FSS", "FMS"]
y = [330, 350, 330, 400, 550, 560]
plt.bar(x, y, color="#4C72B0") # Draw bars
add_labels(x, y) # Overlay labels
plt.title("Faculty Admissions")
plt.xlabel("Faculties")
plt.ylabel("Number of Admissions")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(7)
x = np.random.rand(40) * 10
y = 2.5 * x + np.random.randn(40) * 3
plt.scatter(x, y,
 c='teal', # point fill color
 edgecolors='black', # outline color of each marker
 alpha=0.75) # transparency (0=invisible, 1=solid)
plt.title("Basic Scatter Plot")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [35, 25, 22, 18] # must sum to 100 (or proportions)
colors = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
plt.pie(sizes,
 labels=labels,
 autopct='%1.1f%%', # show percentage inside each slice
 colors=colors,
 startangle=90) # rotate so first slice starts at top
plt.title("Market Share (Pie Chart)")
plt.show()
```

```python
students = ["Amal", "Bimal", "Chamod", "Dinithi", "Eshan"]
scores = [78, 85, 62, 91, 73]

plt.bar(students,scores,linestyle=':')
for i, v in enumerate(scores):
 plt.text(i, # bar x-position (integer index)
 v + 1.5, # just above bar top
 str(v), # label text
 ha='center', # center-align
 fontsize=9)
plt.title("Exam Scores by Student")
plt.xlabel("Student")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
delivery = np.random.normal(35, 8, 250)
plt.hist(delivery,
 bins=18, # 18 equal-width bins
 color='#55A868', # green bars
 edgecolor='black', # visible bin boundaries
 alpha=0.85) # slight transparency
plt.title("Delivery Time Distribution")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
brands = ["BrandX", "BrandY", "BrandZ", "BrandW", "BrandV"]
share = [28, 24, 21, 15, 12]
colors = ['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2']
# ── Pie chart (Matplotlib) ──
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.pie(share, labels=brands, autopct='%1.1f%%',
 colors=colors, startangle=90)
ax1.set_title("Pie Chart")
# ── Bar chart (Matplotlib) ──
bars = ax2.bar(brands, share, color=colors)
for i, v in enumerate(share):
 ax2.text(i, v + 0.4, f"{v}%", ha='center', fontsize=9)
ax2.set_title("Bar Chart (more accurate)")
ax2.set_ylabel("Market Share (%)")
plt.suptitle("Market Share Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```
